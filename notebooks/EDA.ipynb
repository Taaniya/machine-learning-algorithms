This version of seaborn has function `factorplot` for making categorical plots.

```python
# Plotting categorical plot of average survival rate by each of these features SibSp and Parch.

for i,col in enumerate(["SibSp","Parch"]):
    plt.figure(num=i)
    sns.factorplot(x=col,y="Survived",data=df,kind="point",aspect=2)
    plt.show()
```

This is a categorical plot showing the average survival score for each of the categories of features `SibSp` & `Parch` respectively. The point is the average metric and the vertical lines indicate the population size at 95% confidence interval calculated using bootstrap.

Let's see if gender has any affect on the survival rate of the passengers.

```python
# Plotting categorical plot of average survival rate by each of these features SibSp and Parch.

for i,col in enumerate(["SibSp","Parch"]):
    plt.figure(num=i)
    sns.factorplot(x=col,y="Survived",data=df,hue='Sex',kind="point",aspect=2)
    plt.show()
```
